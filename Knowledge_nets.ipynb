```python
import os
import numpy as np
import pandas as pd
import ast
import re
import csv

import nltk
#from nltk.tokenize import PunktSentenceTokenizer
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
ps = PorterStemmer()
lemma = WordNetLemmatizer()

from sklearn.feature_extraction.text import CountVectorizer


import gensim
from gensim import corpora, models
```

```python
df = pd.read_csv('C:/Python37/Replicated-Softmax-Model-master/data/small_set.csv', usecols=['Abstract'],error_bad_lines=False)
df.dropna(inplace=True)
df.drop_duplicates(keep=False,inplace=True) 
#print(df)
```

```python
from collections import defaultdict
print('------- Data cleaning -------')                
stopwords_en = stopwords.words('english')
dat_clean = []
#dat_clean = defaultdict(list)
df = df.apply(lambda row: nltk.word_tokenize(row['Abstract']),axis=1)
#print(df)
#df = df.apply(lambda row: [w.split('.') for w in df if len(w.split('.')) > 1])
#print(df)
#df = df.apply(lambda x: (len(x['df']).split('.')) -1)
df = df.astype(str)
for i in range(len(df)):

    ''' tokenization and punctuation removal '''
    
    
    
    # split words sperated by fullstops
    tmp_doc_split = [w.split('.') for w in df if len(w.split('.')) > 1]
    # flatten list
    tmp_doc_split = [i_sublist for i_list in tmp_doc_split for i_sublist in i_list]
    # clean split words
    tmp_doc_split = [w for w in tmp_doc_split if re.search('^[a-z]+$',w)]
    
    # drop punctuations
    tmp_doc_clean = [w for w in df if re.search('^[a-z]+$',w)]
    tmp_doc_clean.extend(tmp_doc_split)

    ''' stop word removal'''
    tmp_doc_clean_stop = [w for w in tmp_doc_clean if w not in stopwords_en]
    #retain only words with 2 characters or more
    tmp_doc_clean_stop = [w for w in  tmp_doc_clean_stop if len(w) >2]
    
    ''' stemming (using the Porter's algorithm)'''
    tmp_doc_clean_stop_stemmed = [ps.stem(w) for w in tmp_doc_clean_stop]
    dat_clean.append(tmp_doc_clean_stop_stemmed)
    f = open("C:/Python37/Replicated-Softmax-Model-master/data/sample.txt", "w")
    #f.write("Writing into a file")
    f.write(str(dat_clean))
    f.close()
    #print progress
    if i % 100 == 0: print( 'Current progress: ' + str(i) + '/' + str(len(df)) )
```
