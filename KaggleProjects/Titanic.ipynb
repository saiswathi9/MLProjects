```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC 
from sklearn.ensemble import AdaBoostClassifier
from sklearn import preprocessing
import scikitplot as skplt
import matplotlib.pyplot as plt
from sklearn import metrics

le = preprocessing.LabelEncoder()
df = pd.read_csv('C:/Users/sivar/Downloads/all/train.csv', usecols=['PassengerId','Survived','Pclass','Name','Sex','Age','SibSp','Parch','Fare'])
#print(df)
#a = np.array(df) 
#print(a)
#y  = a[:,1]
#print(y)

# mark zero values as missing or NaN
df[['PassengerId','Pclass','Name','Sex','Age','SibSp','Parch','Fare']] = df[['PassengerId','Pclass','Name','Sex','Age','SibSp','Parch','Fare']].replace(0, np.NaN)
# count the number of NaN values in each column
print(df.isnull().sum())
##0 values
'''
PassengerId      0
Survived       549
Pclass           0
Name             0
Sex              0
Age            177
SibSp          608
Parch          678
Fare            15
Cabin          687
Embarked         2
'''

# count the number of NaN values in each column
#print(df.isnull().sum())
#removing rows with missing values
#df.dropna(inplace=True)
df.fillna(df.mean(), inplace=True)
#values = {'Cabin': 0, 'Embarked': 1, 'C': 2, 'D': 3}
#df.fillna(value=values)
print(df.isnull().sum())
for column in df.columns:
    if df[column].dtype == type(object):
        #le = LabelEncoder()
        df[column] = le.fit_transform(df[column])
x = np.column_stack((df.PassengerId,df.Pclass,df.Name,df.Sex,df.Age,df.SibSp,df.Parch,df.Fare)) 
print(x.shape)
y=df.Survived
print(np.unique(y))
print(y.shape)
'''Adaboost'''
abc=AdaBoostClassifier(n_estimators=50,learning_rate=1)
model=abc.fit(x,y)
'''SVM'''
# fitting x samples and y classes 
#clf = SVC(kernel='linear') 
#print("after SVC")
#clf.fit(x, y) 
print("after fit")
test = pd.read_csv('C:/Users/sivar/Downloads/all/test.csv', usecols=['PassengerId','Pclass','Name','Sex','Age','SibSp','Parch','Fare'])
test.fillna(test.mean(), inplace=True)
#test.dropna(inplace=True)
for column in test.columns:
    if test[column].dtype == type(object):
        #le = LabelEncoder()
        test[column] = le.fit_transform(test[column])
xt = np.column_stack((test.PassengerId,test.Pclass,test.Name,test.Sex,test.Age,test.SibSp,test.Parch,test.Fare)) 
xt.shape
#yt = test.Survived
y_pred = model.predict(xt)  
print("Accuracy: ",metrics.accuracy_score(y,y_pred))
#acc = round(clf.score(x, y) * 100, 2)
#print(acc) # 80.02%
#y_pred
submission = pd.DataFrame({
        "PassengerId": test["PassengerId"],
        "Survived": y_pred
    })
submission.to_csv('C:/Users/sivar/Desktop/submission.csv', index=False)

#plotting ROC curve
skplt.metrics.plot_roc_curve(y_pred,y)
plt.show()

#plotting
```
